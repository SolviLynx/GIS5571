```python
import requests
import json
import pprint
import zipfile
```

```python
packages = "https://gisdata.mn.gov/api/3/action/package_list"
groups = "https://gisdata.mn.gov/api/3/action/group_list"
tags = "https://gisdata.mn.gov/api/3/action/tag_list"
```

```python
response = requests.get(tags, auth=('user', 'pass'), verify = False)
```

```python
type(response.content)
```

```python
tags_json = json.loads(response.content)

print(tags_json)
```

```python
pprint.pprint(tags_json)
```

```python
tag = 'transit_routes'
```

```python
base_url = "https://gisdata.mn.gov/api/3/action/package_search?q="
```

```python
package_information_url = base_url + tag
```

```python
print(package_information_url)
```

```python
package_information = requests.get(package_information_url, auth=('user', 'pass'), verify= False)
```

```python
package_dict = json.loads(package_information.content)

pprint.pprint(package_dict)
```

```python
for keys in package_dict['result']['results']:
    pprint.pprint(keys)
```

```python
r = requests.get(keys["resources"][3]["url"])
open('shp_trans_transit_routes.zip', 'wb').write(r.content)
print("extracting the content...")
```

```python
with zipfile.ZipFile('shp_trans_transit_routes.zip', 'r') as zip_ref:
    zip_ref.extractall('C:/Users/rosem/Documents/GIS5571')
    
```
